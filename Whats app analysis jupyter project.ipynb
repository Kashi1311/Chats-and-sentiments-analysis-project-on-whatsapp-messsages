```python
import re
#import Regular Expression
import pandas as pd
```

```python
w = open('WhatsApp Chat with CodeVIT nikalwao.txt','r', encoding = 'utf-8')
```

```python
data = w.read()
print(data)
```

```python
print(type(data))
```

## Now we will seperate date , user name and the user message in different names by using regular expression from the tool regex101.com. ##

```python
pattern = "\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s"
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
##similarly for dates we have ##
dates = re.findall(pattern, data)
dates
```

## Now we will make a pandas data frame . ##

```python
df0 = pd.DataFrame({'users_messages':messages, 'messages_date': dates})
#convert dates into date_type #
df0['messages_date'] = pd.to_datetime(df0['messages_date'], format = '%d/%m/%Y, %H:%M - ')
df0.rename(columns={'messages_date':'date'}, inplace=True)
df0.head()
```

```python
df0.shape
```

## Now we will seperate out the user name and his/her messages. ##

```python
users = []
messages = []
for message in df0['users_messages']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('Group_notification')
        messages.append(entry[0])
        
df0['users'] = users
df0['messages'] = messages
df0.drop(columns = {'users_messages'}, inplace=True)
df0.head(5)
```

## We will split out the date columns into some more sub columns. 

```python
df0['year'] = df0['date'].dt.year
df0.head(5)
```

```python
df0['month'] = df0['date'].dt.month_name()
df0.head()
```

```python
df0['day'] = df0['date'].dt.day
df0.head()
```

```python
df0['hour'] = df0['date'].dt.hour
df0.head()
```

```python
df0['minutes'] = df0['date'].dt.minute
df0.head()
```
